```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/rahul/Desktop/GDSC-ML Tasks/New Dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv')
pd.set_option('display.max_columns', None)
df.head(10)
```

## Data Cleaning:

```python
df.loc[df['TotalCharges']==' ']
```

```python
df1 = df.copy()
```

```python
#Handle missing values in TotalCharges
df1['TotalCharges'] = df1['TotalCharges'].replace(' ', np.nan)
df1['TotalCharges'] = pd.to_numeric(df1['TotalCharges'], errors='coerce')
df1['TotalCharges'] = df1['TotalCharges'].fillna(df1['TotalCharges'].median())
```

```python
df1.drop(['customerID'],axis=1,inplace=True)
```

```python
df1['SeniorCitizen'] = df1['SeniorCitizen'].replace({0: 'No', 1: 'Yes'})
df1.head()
```

```python
df1.to_csv('eda_df.csv',index=False)
```

## Plotting Insights:

#### 1. Gender and Churn:

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df1, x='gender', hue='Churn',palette='Dark2',alpha=0.9)
plt.title('Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

##### Insight:

	The churn rate between male and female customers is very similar, suggesting that gender does not significantly impact the decision to churn.

	This implies that customer retention strategies shouldnt necessarily be gender-targeted, but should rather focus on other attributes like tenure or service quality that influence churn more directly.

#### Solution:
Since gender doesnt play a major role in churn prediction, retention strategies should focus on factors that more strongly correlate with churn, such as pricing models or service-related issues.


#### 2. Tenure and Churn:

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df1, x='tenure', hue='Churn', kde=True, bins=30, palette='Dark2',alpha =0.5,multiple='stack')
plt.title('Churn by Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Count')
plt.show()
```

##### Insight:

	Customers with shorter tenure (less than 6 months) are much more likely to churn, as indicated by the sharp increase in churn early on.

	Customers with longer tenure (beyond 24 months) are less likely to churn, showing that customers who stay past the early months tend to remain loyal.

	This suggests that the first few months are critical in building customer loyalty.

#### Solution:
Implementing early-stage interventions, such as onboarding support, promotions, or personalized offers, could help retain customers in their first six months. This could include increased communication, loyalty programs, or targeted discounts.


#### 3. Monthly Charges and Churn:

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df1, x='Churn', y='MonthlyCharges', palette='Dark2')
plt.title('Churn by Monthly Charges')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```

##### Insight:

	Customers with higher monthly charges are more likely to churn, as shown by the higher distribution of monthly charges among those who have left.

	Customers with lower monthly charges appear more likely to stay, indicating that price sensitivity plays a major role in churn.

#### Solution:
Consider providing customized pricing plans for high-spending customers. Offering discounts, loyalty benefits, or bundled services could reduce churn among those paying higher monthly charges. Reducing unexpected or unnecessary charges could also increase customer satisfaction.


#### 4. Contract Type and Churn:

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df1, x='Contract', hue='Churn', palette='Dark2')
plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.show()
```

##### Insight:

	Customers with month-to-month contracts are significantly more likely to churn compared to those with one or two-year contracts.

	The lower churn rate among long-term contract holders suggests that contract length is a strong retention factor, likely because customers are locked into these agreements and may receive better pricing incentives.

#### Solution:
Encourage more customers to move from month-to-month to longer-term contracts by offering discounts or exclusive benefits. Promotional offers, such as reduced rates for signing up for one-year contracts, can increase the commitment level and reduce churn risk.
