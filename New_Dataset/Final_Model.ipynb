```python
df1 = df.copy()
```

```python
#Handle missing values in TotalCharges
df1['TotalCharges'] = df1['TotalCharges'].replace(' ', np.nan)
df1['TotalCharges'] = pd.to_numeric(df1['TotalCharges'], errors='coerce')
df1['TotalCharges'] = df1['TotalCharges'].fillna(df1['TotalCharges'].median())
```

```python
binary_columns = ['SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
#Replacing the Yes and Nos to 1s and 0s
df1[binary_columns] = df1[binary_columns].replace({'Yes': 1, 'No': 0})
```

```python
#Drop unnecessary columns:
df1.drop(columns=['customerID'], inplace =True)
```

```python
#Label Encoding the contract column:
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Fit and transform the 'Contract' column
df1['Contract'] = le.fit_transform(df1['Contract'])

# Check the encoded values
print(df1['Contract'].unique())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define the columns to OneHotEncode
columns_to_encode = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                     'StreamingTV', 'StreamingMovies', 'MultipleLines', 
                     'gender', 'PaymentMethod', 'InternetService']

df_encoded = pd.get_dummies(df1, columns=columns_to_encode)

# Convert boolean values (True/False) to integers (0/1)
df_encoded = df_encoded.astype(int)
df_encoded.head()
```
