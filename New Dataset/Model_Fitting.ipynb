```python
df1 = df.copy()
```

```python
#Handle missing values in TotalCharges
df1['TotalCharges'] = df1['TotalCharges'].replace(' ', np.nan)
df1['TotalCharges'] = pd.to_numeric(df1['TotalCharges'], errors='coerce')
df1['TotalCharges'] = df1['TotalCharges'].fillna(df1['TotalCharges'].median())
```

```python
# Yes/No - SeniorCitizen, Partner, Dependents, PhoneService, PaperlessBilling, Churn -->convert to 1s and 0s
#Yes/No/No internet service - OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, StreamingTV, StreamingMovies -->OneHotEncode
#Yes/No/No phone service - MultipleLines -->OneHotEncode
#Categorical Ordinal - Contract -->LabelEncode
#Categorical Nominal - Gender, Payment Method, InternetService --> OneHotEncode
#int64 - tenure, MonthlyCharges, TotalCharges --> keep as it is
#Drop - customerId --> Drop as its not necessary
```

```python
binary_columns = ['SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
#Replacing the Yes and Nos to 1s and 0s
df1[binary_columns] = df1[binary_columns].replace({'Yes': 1, 'No': 0})
```

```python
#Drop unnecessary columns:
df1.drop(columns=['customerID'], inplace =True)
```

```python
#Label Encoding the contract column:
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Fit and transform the 'Contract' column
df1['Contract'] = le.fit_transform(df1['Contract'])

# Check the encoded values
print(df1['Contract'].unique())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define the columns to OneHotEncode
columns_to_encode = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                     'StreamingTV', 'StreamingMovies', 'MultipleLines', 
                     'gender', 'PaymentMethod', 'InternetService']

# Initialize the OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)

# Create a ColumnTransformer to apply OneHotEncoder to the specified columns
column_transformer = ColumnTransformer(
    transformers=[('ohe', ohe, columns_to_encode)], remainder='passthrough')

# Fit and transform the data
df_encoded = column_transformer.fit_transform(df1)

# Retrieve the feature names after fitting
encoded_columns = column_transformer.named_transformers_['ohe'].get_feature_names_out(columns_to_encode)

# Convert the transformed data back to a DataFrame with the correct column names
df_encoded = pd.DataFrame(df_encoded, columns= df1.drop(columns_to_encode, axis=1).columns.tolist() + list(encoded_columns))

# Check the encoded DataFrame
df_encoded.head()
```
