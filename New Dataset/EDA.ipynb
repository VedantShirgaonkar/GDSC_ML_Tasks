```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/rahul/Desktop/GDSC-ML Tasks/New Dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv')
pd.set_option('display.max_columns', None)
df.head(10)
```

```python
for column in df.columns:
    val_counts = df[column].value_counts()
    print(f"Column: {column}")
    print(f"Unique values: {val_counts}")
    print("\n")
```

## Data Cleaning:

```python
df.loc[df['TotalCharges']==' ']
```

```python
df1 = df.copy()
```

```python
#Handle missing values in TotalCharges
df1['TotalCharges'] = df1['TotalCharges'].replace(' ', np.nan)
df1['TotalCharges'] = pd.to_numeric(df1['TotalCharges'], errors='coerce')
df1['TotalCharges'] = df1['TotalCharges'].fillna(df1['TotalCharges'].median())
```

```python
df1.drop(['customerID'],axis=1,inplace=True)
```

```python
df1['SeniorCitizen'] = df1['SeniorCitizen'].replace({0: 'No', 1: 'Yes'})
df1.head()
```

## Plotting Insights:

```python
# Insight 1: Gender and Churn
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender', hue='Churn')
plt.title('Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Insight 2: Tenure and Churn
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='tenure', hue='Churn', kde=False, bins=30, palette='Set1')
plt.title('Churn by Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Count')
plt.show()
```

```python
# Insight 3: Monthly Charges and Churn
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='Set1')
plt.title('Churn by Monthly Charges')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```

```python

# Insight 4: Contract Type and Churn
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set1')
plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.show()
```

```python
# Insight 5: Internet Service Type and Churn
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='InternetService', hue='Churn', palette='Set1')
plt.title('Churn by Internet Service Type')
plt.xlabel('Internet Service Type')
plt.ylabel('Count')
plt.show()
```

```python
# Insight 6: Total Charges and Churn
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')  # Convert TotalCharges to numeric, handling errors
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='TotalCharges', hue='Churn', kde=False, bins=30, palette='Set1')
plt.title('Churn by Total Charges')
plt.xlabel('Total Charges')
plt.ylabel('Count')
plt.show()
```
