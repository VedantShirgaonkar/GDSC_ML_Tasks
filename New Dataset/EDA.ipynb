## Plotting Insights:

```python
# Insight 1: Gender and Churn
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender', hue='Churn',palette='Dark2',alpha=0.9)
plt.title('Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#### 1. Gender and Churn

##### Insight: The churn rate between male and female customers appears to be fairly balanced, indicating that gender alone may not be a significant predictor of churn.
##### Explanation: The count plot visualizes the distribution of churn across genders, showing a roughly equal churn rate. This suggests that models like Logistic Regression or Random Forest, which rely on multiple features, may be more appropriate as gender alone won’t be sufficient to predict churn.

```python
# Insight 2: Tenure and Churn
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='tenure', hue='Churn', kde=True, bins=30, palette='Dark2',alpha =0.5,multiple='stack')
plt.title('Churn by Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Count')
plt.show()
```

#### 2. Tenure and Churn

##### •	Insight: Customers with shorter tenure (less than 10 months) are more likely to churn, while those with longer tenure tend to stay.
##### •	Explanation: The histogram of tenure shows a high churn rate among customers who have been with the company for a shorter time. This pattern can suggest that tenure is an important feature for predicting churn. We could use this information to apply models like Decision Trees that can capture these non-linear relationships between tenure and churn.

```python
# Insight 3: Monthly Charges and Churn
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='Dark2')
plt.title('Churn by Monthly Charges')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```

#### 3. Monthly Charges and Churn

##### •	Insight: Higher monthly charges are associated with a higher likelihood of churn.
##### •	Explanation: From the boxplot, we observe that customers who churn tend to have higher monthly charges. This can help in suggesting models such as XGBoost or Support Vector Machines (SVM), which can handle complex relationships between variables like monthly charges and churn.

```python
# Insight 4: Contract Type and Churn
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Contract', hue='Churn', palette='Dark2')
plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.show()
```

#### 4. Contract Type and Churn

##### •	Insight: Customers with month-to-month contracts are more likely to churn compared to those with one or two-year contracts.
##### •	Explanation: The count plot shows a much higher churn rate for month-to-month contracts. This indicates that the type of contract a customer holds could be a strong predictor of churn. This feature could be especially useful in a Random Forest model where categorical data has a strong predictive power.

```python
# Insight 5: Internet Service Type and Churn
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='InternetService', hue='Churn', palette='Dark2')
plt.title('Churn by Internet Service Type')
plt.xlabel('Internet Service Type')
plt.ylabel('Count')
plt.show()
```

#### 5. Internet Service Type and Churn

##### •	Insight: Fiber optic customers are more likely to churn compared to those using DSL or no internet service.
##### •	Explanation: The count plot reveals that customers using fiber optic internet service have a higher churn rate. This suggests that the quality or satisfaction with fiber optic services may need attention. A Gradient Boosting model could be useful here, as it can effectively deal with categorical variables like internet service.

```python
# Insight 6: Total Charges and Churn
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')  # Convert TotalCharges to numeric, handling errors
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='TotalCharges', hue='Churn', kde=False, bins=30, palette='Dark2',alpha=0.5,multiple='stack')
plt.title('Churn by Total Charges')
plt.xlabel('Total Charges')
plt.ylabel('Count')
plt.show()
```

#### 6. Total Charges and Churn

##### •	Insight: Churn is more prevalent among customers with lower total charges, indicating that new customers or those with fewer services may be more likely to leave.
##### •	Explanation: The distribution plot shows that churned customers often have lower total charges. This feature can provide valuable input to models like Logistic Regression, especially if the relationship between total charges and churn is linear, as well as non-linear models like Decision Trees.
