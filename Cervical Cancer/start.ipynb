```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Metadata:
1. Age:

The age of the individual in years.

2. Number of sexual partners:

The total number of sexual partners the individual has had.

3. First sexual intercourse:

The age at which the individual had their first sexual intercourse.

4. Num of pregnancies:

The number of pregnancies the individual has experienced.

5. Smokes:

Whether the individual is a smoker (1 = Yes, 0 = No).

6. Smokes (years):

The number of years the individual has been smoking.

7. Smokes (packs/year):

The average number of cigarette packs smoked per year by the individual.

8. Hormonal Contraceptives:

Whether the individual has used hormonal contraceptives (1 = Yes, 0 = No).

9. Hormonal Contraceptives (years):

The number of years the individual has been using hormonal contraceptives.

10. IUD (Intrauterine Device):

Whether the individual has used an intrauterine device (1 = Yes, 0 = No).

11. IUD (years):

The number of years the individual has used an intrauterine device.

12. STDs:

Whether the individual has had any sexually transmitted diseases (STDs) (1 = Yes, 0 = No).

13. STDs (number):

The total number of STDs the individual has had.

14. STDs: Time since first diagnosis:

The number of years since the individual was first diagnosed with an STD.

15. STDs: Time since last diagnosis:

The number of years since the individual was last diagnosed with an STD.

16. STDs:condylomatosis:

Whether the individual has been diagnosed with condylomatosis, a type of genital wart (1 = Yes, 0 = No).

17. STDs:cervical condylomatosis:

Whether the individual has been diagnosed with cervical condylomatosis (1 = Yes, 0 = No).

18. STDs:vaginal condylomatosis:

Whether the individual has been diagnosed with vaginal condylomatosis (1 = Yes, 0 = No).

19. STDs:vulvo-perineal condylomatosis:

Whether the individual has been diagnosed with condylomatosis in the vulva or perineal areas (1 = Yes, 0 = No).

20. STDs:syphilis:

Whether the individual has been diagnosed with syphilis, an STD (1 = Yes, 0 = No).

21. STDs:pelvic inflammatory disease:

Whether the individual has been diagnosed with pelvic inflammatory disease (1 = Yes, 0 = No).

22. STDs:genital herpes:

Whether the individual has been diagnosed with genital herpes (1 = Yes, 0 = No).

23. STDs:molluscum contagiosum:

Whether the individual has been diagnosed with molluscum contagiosum, a viral infection (1 = Yes, 0 = No).

24. STDs:AIDS:

Whether the individual has been diagnosed with AIDS (1 = Yes, 0 = No).

25. STDs:HIV:

Whether the individual has been diagnosed with HIV, the virus that can lead to AIDS (1 = Yes, 0 = No).

26. STDs:Hepatitis B:

Whether the individual has been diagnosed with Hepatitis B, a liver infection caused by the Hepatitis B virus (1 = Yes, 0 = No).

27. STDs:HPV:

Whether the individual has been diagnosed with human papillomavirus (HPV), a common STD (1 = Yes, 0 = No).

28. STDs: Number of diagnosis:

The total number of STD diagnoses the individual has received.

29. Dx:Cancer:

Whether the individual has been diagnosed with cancer (1 = Yes, 0 = No).

30. Dx:CIN:

Whether the individual has been diagnosed with cervical intraepithelial neoplasia (CIN), which is a potential precursor to cervical cancer (1 = Yes, 0 = No).

31. Dx:HPV:

Whether the individual has been diagnosed with HPV (1 = Yes, 0 = No).

32. Dx:

General diagnosis indicating whether the individual has been diagnosed with any disease (1 = Yes, 0 = No).

33. Hinselmann:

Whether the individual has undergone a Hinselmann test (1 = Positive result, 0 = Negative result). This test looks for abnormalities in cervical tissue.

34. Schiller:

Whether the individual has undergone a Schiller test (1 = Positive result, 0 = Negative result). This test uses iodine to identify abnormal cells in the cervix.

35. Citology:

Whether the individual has undergone a Pap smear cytology test (1 = Positive result, 0 = Negative result).

36. Biopsy:

Whether the individual has had a biopsy for cervical abnormalities (1 = Positive result, 0 = Negative result).

```python
df1 = df.copy()
```

```python
df1['Number of sexual partners'] = pd.to_numeric(df1['Number of sexual partners'], errors='coerce')
df1['Number of sexual partners'] = df1['Number of sexual partners'].fillna(df1['Number of sexual partners'].median())
```

```python
df1['First sexual intercourse'] = pd.to_numeric(df1['First sexual intercourse'], errors='coerce')
df1['First sexual intercourse'] = df1['First sexual intercourse'].fillna(df1['First sexual intercourse'].median())
```

```python
df1['Num of pregnancies'] = pd.to_numeric(df1['Num of pregnancies'], errors='coerce')
df1['Num of pregnancies'] = df1['Num of pregnancies'].fillna(df1['Num of pregnancies'].median())
```

```python
df1['Smokes'] = pd.to_numeric(df1['Smokes'], errors='coerce')
df1['Smokes'] = df1['Smokes'].fillna(df1['Smokes'].mode()[0])
```

```python
df1.Smokes.value_counts()
```

```python
df.loc[(df['Smokes']=='?') & (df['Smokes (years)']=='?')].shape
```

```python
df.loc[(df['Smokes (packs/year)']=='?') & (df['Smokes']=='?')].shape
```

```python
#This shows that the same rows are missing from the 3 columns of Smokes
#Therefore, replace all the corresponding values with 0
df1['Smokes (years)'] = pd.to_numeric(df1['Smokes (years)'], errors='coerce')
df1['Smokes (years)'] = df1['Smokes (years)'].fillna(0)

df1['Smokes (packs/year)'] = pd.to_numeric(df1['Smokes (packs/year)'], errors='coerce')
df1['Smokes (packs/year)'] = df1['Smokes (packs/year)'].fillna(0)
```

```python
#There is a large no of missing values for hormonal contraceptives and hence we cannot directly impute it. 
#First we need to explore its relationships with other variables

# Create a figure with multiple subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(11, 11))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# List of variables to plot against 'Hormonal Contraceptives'
variables = ['Age', 'Number of sexual partners', 'Num of pregnancies', 
             'First sexual intercourse', 'Smokes','STDs']

# Create scatter plots
for ax, var in zip(axes, variables):
    sns.scatterplot(x=df[var], y=df['Hormonal Contraceptives'], ax=ax)
    ax.set_title(f'Hormonal Contraceptives vs {var}')
    ax.set_xlabel(var)
    ax.set_ylabel('Hormonal Contraceptives')

# Remove the last empty subplot if there's an uneven number of plots
if len(axes) > len(variables):
    fig.delaxes(axes[-1])

# Adjust layout
plt.tight_layout()
plt.suptitle('Relationship of Hormonal Contraceptives with Other Variables', y=1.02)
plt.show()
```

```python
#Therefore there is no relationship of Hormonal Contraceptive with other variables and hence we can directly impute it with mode
df1['Hormonal Contraceptives'] = pd.to_numeric(df1['Hormonal Contraceptives'], errors='coerce')
df1['Hormonal Contraceptives'] = df1['Hormonal Contraceptives'].fillna(df1['Hormonal Contraceptives'].mode()[0])
```

```python
df1['Hormonal Contraceptives (years)'] = pd.to_numeric(df1['Hormonal Contraceptives (years)'], errors='coerce')
df1['Hormonal Contraceptives (years)'] = df1['Hormonal Contraceptives (years)'].fillna(df1['Hormonal Contraceptives (years)'].median())
```

```python
#All the missing values of different types of STD are from the same row as 'STD'. Hence we can replace all of them by 0
std_cols = ['STDs (number)', 'STDs:condylomatosis',
       'STDs:cervical condylomatosis', 'STDs:vaginal condylomatosis',
       'STDs:vulvo-perineal condylomatosis', 'STDs:syphilis',
       'STDs:pelvic inflammatory disease', 'STDs:genital herpes',
       'STDs:molluscum contagiosum', 'STDs:AIDS', 'STDs:HIV',
       'STDs:Hepatitis B', 'STDs:HPV']
for col in std_cols:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')
    df1[col] = df1[col].fillna(0)
```
